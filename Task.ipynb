```python
#2. Find the length of the longest continuous increasing subsequence from an unsortes array of integers.

def find_length_of_LCIS(numbers):
    if not numbers:
        return 0

    max_length = 1
    current_length = 1

    for i in range(1,len(numbers)):
        if numbers[i] > numbers[i - 1]:
            current_length += 1
        else:
            max_length = max(max_length, current_length)
            current_length = 1

    # Final comparison to cover the case when the longest increasing sequence is at the end
    max_length = max(max_length, current_length)

    return max_length

numbers = input("Enter the numbers seperated by space: ").split()
print(find_length_of_LCIS(numbers))
```

```python
#3. Arrange the list of non negative numbers such that they form the largest number.

def largest_number(nums):
    # Convert integers to strings
    nums_str = list(map(str, nums))

    # Define a custom comparator
    def compare(x, y):
        return (y + x > x + y) - (y + x < x + y)

    # Sort the strings using the custom comparator
    nums_str.sort(key=lambda x: x * 10, reverse=True)

    # Join the sorted list into a single string
    largest_num = ''.join(nums_str)

    # Edge case: when the sorted numbers are all zeros
    if largest_num[0] == '0':
        return '0'

    return largest_num


nums = input("Enter the numbers seperated by space: ").split()
print(largest_number(nums))
```
